```python
#Imports
import pandas as pd
import numpy as np

data = pd.read_csv('data.csv')
print(data.shape)
data.head()
```

```python
#Splitting the data into features and labels
features = data.iloc[:, 1:]
labels = data.iloc[:, 0]
```

```python
#Splitting the data in training and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, train_size = 0.8, test_size = 0.2, 
                                                    random_state = 0)
```

```python
#Building the neural network model
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping

model = Sequential()

model.add(Dense(256, activation = 'relu', input_shape=(X_train.shape[1],)))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(1))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')
```

```python
#Fitting the model
model.fit(X_train, y_train, validation_split = 0.2, epochs = 1000, batch_size = 50, callbacks = [EarlyStopping(patience = 10)])
```

```python
#Evaluating the model
score = model.evaluate(X_test, y_test, verbose = 0)
print("Mean Squared Error is {}".format(score))
```
